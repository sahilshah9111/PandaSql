```python
query1 = "SELECT SUM(Quantity) from drinks WHERE continent =='AS' AND Drink_Name == 'beer_servings' GROUP BY continent"
```

```python
query3 = "SELECT SUM(Quantity), continent from drinks WHERE Drink_Name IN ('spirit_servings', 'wine_servings') GROUP BY continent HAVING SUM(Quantity)>2000"
s_w = sqldf(query3)
```

Task 4

```python
query4 = "select continent, \
          max(case when Drink_Name = 'beer_servings' then Quantity end) as Beer, \
          max(case when Drink_Name = 'spirit_servings' then Quantity end) as Spirit \
from drinks \
group by continent \
having max(case when Drink_Name = 'beer_servings' then Quantity end) > max(case when Drink_Name = 'spirit_servings' then Quantity end)"
```

```python
b_s = sqldf(query4)
b_s
```
